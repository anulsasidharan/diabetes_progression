```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
pd.read_csv("diabetes.csv")
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
x = df.drop(columns="Y")
```

```python
x.head()
```

```python
y = df["Y"]
```

```python
y
```

```python
# Split the test and train datasets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
# Standardize the data to do the ML predection
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.fit_transform(x_test)
```

```python
x_train_scaled
```

```python
ridge_model = Ridge()
lasso_model = Lasso()
```

```python
# Pass the training dataset to Ridge and lasso model 
ridge_model.fit(x_train_scaled, y_train)
```

```python
lasso_model.fit(x_train_scaled, y_train)
```

```python
# Predect the diabetes progression predection
# Ridge
# Pass the x_test_scaled data to predict the outcome.
y_pred_ridge = ridge_model.predict(x_test_scaled)
```

```python
# Predect the diabetes progression predection
# Lasso
# Pass the x_test_scalled data to predect the outcome.
y_pred_lasso = lasso_model.predict(x_test_scaled)
```

```python
# Let's find out the accuracy od the model we have created by uding Mean Square Error
# with respect to the y_test and y_pred data 
mean_squared_error(y_test, y_pred_ridge)
```

```python
mean_absolute_error(y_test, y_pred_ridge)
```

```python
r2_score(y_test, y_pred_ridge)
```

#####  This means out prediction is 44 % accurate. 


```python
# Let's create pickle file for saving the model. 
# Ridge Model Pickle file
import pickle
with open ("diabetes_ridge_model.pkl", "wb") as file:
    pickle.dump((ridge_model, scaler), file)
```

```python
# Let's create pickle file for saving the lasso model.
# lasso model pickle file
with open ("diabetes_lasso_model.pkl", "wb") as file:
    pickle.dump((lasso_model, scaler), file)
```
